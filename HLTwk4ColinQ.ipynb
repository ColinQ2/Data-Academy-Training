```python
"""
Task 1 - Write a program that allows you to enter 4 numbers and store them in a file called numbers
"""
MaxNum = 4                              #Specify numbers to ask
x = [0] * MaxNum                        #Assign array
for y in range(MaxNum):                 #Iterate as needed
    x[y] = input("Specify a number")    #Collect input in array

#print (x)    

Task1_File = open("Week4Task1_Store4Numbers.txt","w") #Create txt file
Task1_File.write(str(x))                                #Write array from input gathering
Task1_File.close()                                      #Close file
```

```python
"""
Task 2 - Write a program to ask a student for their percentage mark and convert this to a grade. The conversion will be done in a function called mark_grade
Extension to task 2
Ask the user for their target grade and print this with their mark
If their target grade > exam grade display a suitablemessage
If their target grade = exam grade display a suitablemessage
If their target grade < exam grade display a suitablemessage
"""

#Convert percentage to grade
def mark_grade(Score):
    if Score >= 90 and Score <= 100:
        GradeAchieved = "A*" 
    elif Score >= 80 and Score < 90:
            GradeAchieved = "A"
    elif Score >= 70 and Score < 80:
            GradeAchieved = "B"
    elif Score >= 60 and Score < 70:
            GradeAchieved = "C"
    elif Score >= 50 and Score < 60:
            GradeAchieved = "D"
    elif Score >= 0 and Score < 50:
            GradeAchieved = "Fail"
    return GradeAchieved            

ScoreX = int(input("What score did you get in the test? 0-100"))    #Gather percentage score
TargetGrade = input("Which grade were you aiming for?")             #Gather target grade
ExamGrade = mark_grade(ScoreX)                                      #Convert score to grade
GradeList = ["Fail","D","C","B","A","A*"]                           #Define list of grades for indexing
ExamGradeInd = GradeList.index(ExamGrade)                           #Get index position of grade
TargetGradeInd = GradeList.index(TargetGrade)                       #Get index position of target grade
#print(ExamGradeInd, TargetGradeInd)

if TargetGradeInd > ExamGradeInd:
    print("Unfortunately a score of", ScoreX, "equates to the grade", ExamGrade, "which is below your target of", TargetGrade)
if TargetGradeInd == ExamGradeInd:
    print("Well done, you've score", ScoreX, "which equates to the grade", ExamGrade, ", just as you were aiming for.")
if TargetGradeInd < ExamGradeInd:
    print("Congratulations, your score of", ScoreX, "equates to the grade", ExamGrade, "which is above your target of", TargetGrade)                                                 
```
