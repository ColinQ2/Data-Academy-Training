```python
"""
Task 1
Write a program that does the following:
a) Stores a random number (1-10) in a variable – see hint below.
b) Asks a user for their name and stores this in a variable.
c) Asks a user to guess the number between 1 and 10.
d) Tells the user whether they have guessed correctly.

Solution
1. import random library
2. Ask python to pick a random number between 1-10
3. Ask user what is their name
4. Ask user to guess a number
5. Create an IF statement to determine if the number the user guessed is the same as the random number generated by python 
"""
import random
RandomNum = random.randint(1,10)
#print(RandomNum)
UserName = input("What is your name?")
UserGuess = input("Guess a number")
#UserGuess = RandomNum
if UserGuess == RandomNum:
    print("Well done" , UserName + ", you guessed" , UserGuess , "which is correct.")
else:
    print("Sorry" , UserName + ", but you guessed" , UserGuess , "which is incorrect. The correct answer guess was" , RandomNum)

```

```python
"""
Task 2
Write a program that asks a user for their favourite number between 1 and 100 and then tells them a joke based on the number. You should use a minimum of 3 jokes.

Solution
1. Ask the user to pick a number
2. Create an IF statement which uses logical operators which prints out a joke based on the number the user picked
"""
NumX= int(input("Please select a number between 1 and 100"))
if NumX >=1 and NumX <=5:
    print("Why did the user input" , str(NumX) + "? \n Because the script told it to.")
elif NumX >=6 and NumX <=9:
    print("Why is 6 afraid of 7? \n Because 7 8 9.")
elif NumX >=10 and NumX <=100:
    print("Why did the coder bucket" , str(NumX) , "into the last joke range? \n Because they ran out of ideas.") 
```

```python
"""
Task 3
Write a program that allows user to enter their favourite starter, main course, dessert and drink.Concatenate these and output a message which says – “Your favourite meal is  ………with a glass of….”

Solution
1. Ask the user what their favourite starter is
2. Ask the user what their favourite main is
3. Ask the user what their favourite dessert is
4. Ask the user what their favourite drink is
5. Concatenate together the users input into a a string and print out
"""
Starter = input("What your favourite starter?")
Main = input("What your favourite main?")
Dessert = input("What your favourite dessert?")
Drink = input("What your favourite drink?")
MSG1 = "Your favourite meal is" , Starter , Main  , Dessert , "with a glass of" , Drink
MSG2 = "Your favourite meal is " + Starter + ", " + Main + ", " + Dessert + " with a glass of " + Drink
print(MSG1)
print(MSG2) #Cleanest print

print("Your favourite meal is" , Starter , ","  ,  Main  ,  ","  , Dessert , "with a glass of" , Drink)
print("Your favourite meal is " + Starter + " ," + Main + ", " + Dessert + " with a glass of " + Drink) #Cleanest print
```

```python
"""
Task 4
A motorbike costs £2000 and loses 10% of its value every year. Using a loop, print the value of the bike every following year until it falls below £1000.

Solution
1. Store the value of the bike against a variable
2. Create a while loop which subtracts 10% of the bikes value until the value reaches £1000
"""
Initial = 2000
Deprec = 0.1
CurValue = Initial
Tick = 0
while CurValue > 1000:
    print("In year" , Tick , "the bike is worth" , CurValue)
    CurValue = CurValue - (CurValue * Deprec)
    Tick += 1
    
```

```python
"""
Task 5
Write a program which will ask for two numbers from a user. Then offer a menu to the user giving them a choice of operator:e.g. – Enter “a” if you want to add“b” if you want to subtract.
Include +, -, /, *, ** square (to the power of). Once the user has selected which operator they wish to use, perform the calculation.

Solution
1. Ask the user to pick a number
2. Ask the user to pick a second number
3. Ask the user to pick an operator
4. Create an IF statement to carry out the math operation using the users choice of numbers and choice of operator
"""
import operator

operator_dict = {
    '+': operator.add,
    '-': operator.sub,
    '/': operator.truediv,
    '*': operator.mul,
    '**': operator.pow
}
Num1 = input("Pick a number")
Num2 = input("Pick another number")
OpX = input("Pick an operator (+, -, /, *, **")
Output = operator_dict[OpX](int(Num1), int(Num2))
print(Output)
```
